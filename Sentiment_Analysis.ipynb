# **Integer-Encoding Sentiment analysis using SimpleRNN**

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,X_test),(y_train,y_test)=imdb.load_data()
```

```python
X_train #Already tokenized + Encoded data
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_train[0])
```

```python
len(X_train.max())
```

```python
from keras.utils import pad_sequences

# Correct the swapped variables before padding
# X_test currently holds training labels, y_train currently holds test sequences.
# We need X_test to hold test sequences for padding.
temp_X_test = X_test
X_test = y_train
y_train = temp_X_test # This also correctly reassigns y_train to hold training labels

X_train=pad_sequences(X_train,padding='post',maxlen=152)
X_test=pad_sequences(X_test,padding='post',maxlen=152)
```

```python
X_train.shape
```

```python
model=Sequential()
model.add(SimpleRNN(32,input_shape=(152,1),return_sequences=False)) #Here false as we only wants last timestep output
#If want output from each node at there only , ex.:sometimes required during like: name entity recognition(we need or not),machine translation(1-scentence into another)
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

# **Sentiment Analysis using Embeddings**

```python
from keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,X_test),(y_train,y_test)=imdb.load_data()
```

```python
# Correct the swapped variables before padding
# X_test currently holds training labels, y_train currently holds test sequences.
# We need X_test to hold test sequences for padding.
temp_X_test = X_test
X_test = y_train
y_train = temp_X_test # This also correctly reassigns y_train to hold training labels

X_train=pad_sequences(X_train,padding='post',maxlen=152)
X_test=pad_sequences(X_test,padding='post',maxlen=152)
```

```python
X_train.shape
```

```python
model=Sequential()
model.add(Embedding(25000,output_dim=2,input_length=152))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

# Explicitly build the model to ensure model.summary() displays parameters
# The input shape to the model is (batch_size, sequence_length)
model.build(input_shape=(None, 152))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
